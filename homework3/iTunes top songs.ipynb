```python
df = pandas.read_csv('iTunes top songs.csv', header=None, names=['ranking', 'name', 'artist','genre', 'time', 'price'])
```

```python
df['genre']
```

```python
word_count = {}
```

```python
for word in df['genre']:
    if word not in word_count:
        word_count[word] = 1
    else:
        word_count[word] = word_count[word] + 1             
```

```python
word_count
```

```python
pandas.Series(word_count)
```

```python
pandas.Series(word_count).to_frame().reset_index()
```

```python
from wordcloud import WordCloud
wc = WordCloud()
```

```python
w = wc.generate_from_frequencies(word_count)
```

```python
import matplotlib.pyplot as plt
plt.imshow(w)
plt.axis('off')
#By counting the frequency of each genre of songs,
#the Word Cloud chart below shows that Pop, Country, Hip-Hop/Rap are the top 3 popular genres among iTunes users.
#The Word Cloud puts the most important genres in the center, but some smaller ones can fill in the gaps.
```

```python
df['artist'].value_counts()
```

```python
def cleaning(x):
    if x not in ['Artist: Cardi B','Artist: Imagine Dragons','Artist: Post Malone','Artist: Migos','Artist: Brett Young','Artist: Camila Cabello','Artist: XXXTENTACION','Artist: BTS','Artist: Lauv','Artist: Rich The Kid','Artist: Justin Timberlake','Artist: Ed Sheeran','Artist: Kane Brown','Artist: Chris Stapleton','Artist: Drake','Artist: Dua Lipa']:
        return False
    else:
        return True
s_artist = df[df['artist'].apply(cleaning)]['artist'].value_counts()
s_artist
```

```python
plt.plot(s_artist.values, s_artist.index)
```

```python
s_artist = s_artist.sort_index()
```

```python
plt.figure(figsize=(10, 5))
s_artist = s_artist.sort_index()
ascending = False
plt.barh(s_artist.index,s_artist.values, color='#75d2e5', edgecolor='#758be5')
plt.yticks(s_artist.index)
plt.xticks(s_artist.values)
ticks_index = range(0, 55, 5)
plt.title('Top artists')
plt.xlabel('Number of songs listed')
plt.ylabel('Artists')
#Cardi B top the list with 4 selected songs,
#followed by Rich The Kid and Imagine Dragons, with 3 songs listed respectively.
```
